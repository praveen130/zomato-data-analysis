```python
import os
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('/home/praveen/zomato.csv',encoding="ISO-8859-1")
df.head()
```

```python
country = pd.read_excel('/home/praveen/Country-code.xlsx')
countary.head()
```

```python
data = pd.merge(df, country, on='Country Code')
data.head()
```

```python
data['Country'].value_counts().plot(kind='bar',figsize=(10,6))
plt.title('restaurant by country',fontsize=16)
plt.xlabel('country')
plt.ylabel('count')
for idx,val in enumerate(data['Country'].value_counts()):
    plt.text(x=idx - 0.3,y=val+200,s=val,color='red',fontsize=12)
plt.show()
```

```python
data_City = data[data['Country'] =='India']
Total_city =data_City['City'].value_counts()
Total_city.plot.bar(figsize=(14, 7),fontsize=14)
plt.title('Restaurants by City', fontsize=30)                                             
plt.xlabel('City', fontsize=20)
plt.ylabel('No of Restaurants', fontsize=20)
plt.show()
```

```python
data[data['Country']=='India']['Has Online delivery'].value_counts().plot(kind='bar')
plt.title('Online order',fontsize=20)
plt.xlabel('Country',fontsize=30)
plt.ylabel('count',fontsize = 20)
plt.show()
```

```python
data[data['Country']=='India']['Has Online delivery'].value_counts().plot(kind='bar')
plt.title('Online order - Not India',fontsize=20,color = 'navy')
plt.xlabel('Country',fontsize=30)
plt.ylabel('count',fontsize = 20)
plt.show()
```

```python
sns.barplot(data['Cuisines'].value_counts().index[:15], data['Cuisines'].value_counts()[:15])
plt.xlabel('Cuisines', fontsize=20)
plt.ylabel('Number of Resturants', fontsize=20)
plt.title('Top 15 Cuisines on Zomato', fontsize=30)
plt.xticks(rotation = 90)
plt.show()
```

```python
country_average = data.groupby('Country')['Aggregate rating'].mean().reset_index()

plt.figure(figsize=(10,5))
plt.xlabel('Country', fontsize=30)
plt.ylabel('Average Ratings', fontsize=30)
plt.title('Average Ratings on Countries', fontsize=30)
plt.bar(country_average['Country'], country_average['Aggregate rating'])
```

```python
plt.hist(data[data['Country'] == 'India']['Aggregate rating'],bins = 20)
plt.xlabel('Ratings',fontsize = 15)
plt.ylabel('No. of Ratings',fontsize= 15)
plt.show()
```

```python
data[data['Aggregate rating'] == 0]['Votes'].value_counts().plot(kind = 'bar',figsize =(7,7))
plt.xlabel('Ratings')
plt.ylabel('Number of people who have give a rating')
plt.show()
```
